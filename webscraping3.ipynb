```python
Titles=[]
price=[]
no_rating=[]
delivery=[]
Ratings=[]
```

```python
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
titles=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
for i in titles:
    if i.text is None :
        Titles.append("--")
    else:
        Titles.append(i.text)
        
prices_tag=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
for i in prices_tag:
    if i.text is None:
        price.append("--")
    else:
        price.append(i.text)
    
    
    
rating_tag=driver.find_elements_by_xpath(r"//span[@class='a-icon-alt']")
for i in rating_tag:
    if i.text is None:
            no_rating.append("--")
    else:
        no_rating.append(i.text)
    
delivery_tag = driver.find_elements_by_xpath(r"//span[@class='a-text-bold']")
for i in delivery_tag:
    if i.text is None:
        delivery.append("--")
    else:
        delivery.append(i.text)
    
    
    
                 
```
